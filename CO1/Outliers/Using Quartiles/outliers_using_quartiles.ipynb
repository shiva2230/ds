```python

from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import numpy as np

file_path = '/content/drive/My Drive/Data_science lab/6.Outliers/weightdata.csv'
df = pd.read_csv(file_path)  # Corrected the 'pd.readfile' to 'pd.read_csv'

def fnctn(data, col, a=1.5):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - a * IQR  # Corrected variable name 'lowwerlimit' to 'lower_limit'
    upper_limit = Q3 + a * IQR  # Corrected variable name 'upperlimit' to 'upper_limit'
    return data[(data[col] >= lower_limit) & (data[col] <= upper_limit)]  # Fixed logical operator '&' and variable names

columns = ['Index', 'Weight(Pounds)']

for column in columns:
    df = fnctn(df, column)

output = '/content/drive/My Drive/Data_science lab/6.Outliers/output-Sheet1.csv'
df.to_csv(output, index=False)

drive.mount("/content/drive", force_remount=True)
```
